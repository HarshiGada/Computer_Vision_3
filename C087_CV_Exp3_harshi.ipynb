```python
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/girl.png')
```

```python
image.shape
```

```python
# The size of image is multiple of 8 hence no resizing of image
```

```python
img1 = image
img2=resize(img1,(648,648))
img3=img1+30
```

```python
plt.subplot(2, 2, 1)
plt.imshow(img1)
plt.title("Image 1-Original")
plt.subplot(2, 2, 2)
plt.imshow(img2)
plt.title("Image 2-Resized")
plt.subplot(2, 2, 3)
plt.imshow(img3)
plt.title("Image 3-Added Intensity")

```

```python
fd1,hog_img1=hog(img1,orientations=9,pixels_per_cell=(16,16),cells_per_block=(8,8),visualize=True,channel_axis=-1)
fd2,hog_img2=hog(img2,orientations=9,pixels_per_cell=(16,16),cells_per_block=(8,8),visualize=True,channel_axis=-1)
fd3,hog_img3=hog(img3,orientations=9,pixels_per_cell=(16,16),cells_per_block=(8,8),visualize=True,channel_axis=-1)
```

```python
#we are applying hog function on image 1,2 rows ,2 columns to make a block,each cell is 8x8,see image representation of hog,consider all axis,fd1 is feature vectior,hog_img1 is image,9 bin histogram
```

```python
print(len(fd1))
print(len(fd2))
```

```python
plt.subplot(2, 2, 1)
plt.imshow(hog_img1,cmap='gray')
plt.title("Image 1-Original")
plt.subplot(2, 2, 2)
plt.imshow(hog_img2,cmap='gray')
plt.title("Image 2-Resized")
plt.subplot(2, 2, 3)
plt.imshow(hog_img3,cmap='gray')
plt.title("Image 3-Added Intensity")
```

```python
# if hog values are small they are not shown,take bottom 20% to 90% to make it visible,using numpy to enhance the image, not changing the feature descriptor
```

```python
p2_1_1,p90_1_1=np.percentile(hog_img1,(2,90))
p2_1_2,p90_1_2=np.percentile(hog_img2,(2,90))
p2_1_3,p90_1_3=np.percentile(hog_img3,(2,90))
```

```python
hog_img1_e=exposure.rescale_intensity(hog_img1,in_range=(p2_1_1,p90_1_1))
hog_img2_e=exposure.rescale_intensity(hog_img2,in_range=(p2_1_2,p90_1_2))
hog_img3_e=exposure.rescale_intensity(hog_img3,in_range=(p2_1_3,p90_1_3))
```

```python
plt.subplot(2, 2, 1)
plt.imshow(hog_img1_e,cmap='gray')
plt.title("Image 1-Original")
plt.subplot(2, 2, 2)
plt.imshow(hog_img2_e,cmap='gray')
plt.title("Image 2-Resized")
plt.subplot(2, 2, 3)
plt.imshow(hog_img3_e,cmap='gray')
plt.title("Image 3-Added Intensity")
```

On the Girl Image, the Hog is applied to identify the edges of the given image.

For the Number of pixels=(8,8), no of cells per block (2,2), length of Hog vector=215000, if size of the cell is increased to 32,32 then Hog of the image shows boundary of the object.

At some places Hog of the image has black portion/patches, this is because there is no change in intensity along x and y axis in that patch, which is beacause this patch does not have edges.
